# 클린 코드의 의미
프로그래밍 언어의 진정한 의미는 아이디어를 다른 개발자에게 전달하는 것.
- 클린코드인지 판단하는 기준은 다른 엔지니어가 코드를 읽고 유지보수 할 수 있는지에 달려있음.
- 이 책을 통해 파이썬 주요 개념 이해하고 좋은코드, 나쁜코드 차이점 확인한 뒤 훌륭한 코드와 좋은 아키텍쳐 특징 식별해 자신만의 정의를 하기...

# 클린 코드의 중요성
유지보수성 향상, 기술부채 감소, 애자일 개발을 통한 효과적 작업 진행, 성공적인 프로젝트 관리를 위해 중요함
- 유지보수성 향상 -> 민첩한 개발과 지속적 배포
- 기술부채란? : 나쁜 결정이나 적당한 타협의 결과로 생긴 소프트웨어적 결함
    - 현재에서 과거의 방향으로 생각해 보는 것 : 현재 직면한 코드가 과거의 잘못된 코드 때문에 생긴 건 아닐까?
    - 현재에서 미래의 방향으로 생각해 보는 것 : 현재 문제를 적당히 때워 지름길로 간다면 미레에 어떤 문제가 생길까?
    - 코드는 지금 바꾸는 것 보다 미래에 변경하는게 더 어려움. 정말 부채!! 이자를 유발함.
    - 장기적이고 근본적인 문제를 내포하고 있음. 지금 당장은 문제가 안 되지만 언젠가는 프로젝트의 돌발변수가 될 수 있음

## 클린코드에서 코드 포매팅의 역할
클린코드란?
- 코딩표준, 포매팅, 린팅 도구나 다른 검사 도구를 사용한 코드 레이아웃 걸정과 같은 것 그 이상의 것.
- **품질 좋은 소프트웨어를 개발하고, 견고하고 유지보수가 쉬운 시스템을 만들고, 기술부채를 회피하는 것.**
- PEP-8이나 코딩스타일과 관련 없음.. 그 이상의 것. 유지보수성이나 소프트웨어 품질에 관한 것.
- 그러나 코드를 올바르게 포매팅하는 것도 작업을 효율화 하기 위해 중요함

## 프로젝트 코딩 스타일 가이드 준수
코딩 가이드라인
- 품질 표준을 지키기 위해 프로젝트에서 따라야만 하는 최소한의 요구사항
- 코드 일관성 중요
- 파이썬이 따라야 하는 코딩 스타일 PEP-8   

PEP-8 특징
- 검색 효율성(Grepability) : 코드에서 토큰을 grep(텍스트 검색 명령어) 할 수 있는 기능.
     - 키워드 인자(keyword argument)에 앖을 할당할 때는 띄어쓰기를 허용하지는 않지만, 변수에 값을 할당할 때는 띄어쓰기를 사용하도록 권고.
- 일관성 : 코드가 일정한 포맷을 가지면 훨씬 쉽게 읽을 수 있음. 코드 레이아웃, 문서화, 이름 작명 규칙 통일
- 코드 품질 : 코드를 구조화.



# Docstring과 어노테이션
파이썬 코드 안에 직접 문서화를 하는 방법. 주석은 가급적 피하고 문서화를 통해 데이터타입 설명하고 예제 제공.
## Docstring
- 소스코드에 포함된 문서(documentation). 
- **docstring**은 기본적으로 리터럴 문자열.
- 로직의 일부분을 먼서화하기 위해 코드 어딘가에 배치됨.
- 이유가 아닌 설명     


코드에 주석(comment)을 다는 것은 나쁜 습관
- 주석은 코드로 아이디어를 제대로 표현하지 못했음을 나타내는 것. -> 설명이 필요하면 나쁜코드
- 오해의 소지가 있음. 방금 수정한 코드 근처의 주석이 현재와 일치하지 않으면 또다른 위험 초래함.     


docstring이 필요한 이유
- docstring은 주석을 다는 게 아닌 코드의 특정 컴포넌트(모듈, 클래스, 메서드 또는 함수)에 대한 문서화.
- 파이썬이 동적타이핑을 하기 때문에(함수의 파라미터 값으로 타입을 체크하거나 강요하지 않기 때문에) 함수의 이름과 파라미터 명이 충분히 설명적이면 좋음.
- 런타임 중에 접근하고 소스코드에서 docstring 내용을 추가하거나 컴파일 하는 것이 가능. Sphinx(스핑크스)실행하면 프로젝트 문서화를 위한 기본 골격 만들어줌.
- 지속적으로 수작업 해야 함. 코드 변경 시 위키(wikji), 사용자 매뉴얼, README파일, docstring등 관련 내용 업데이트 중요.    

## 어노테이션
- 코드 사용자에게 함수 인자로 어떤 값이 와야하는지 **힌트**를 주는 것.
- 변수의 예상 타입을 지정할 수 있음.

## 어노테이션은 docstring을 대체하는 것?
네!!
- docstring과 어노테이션은 서로 보완적인 개념.
- docstring에 포함된 정보의 일부는 어노테이션으로 이동시킬 ㅅ ㅜ있는 것이 사실.

## 기본 품질 향상을 위한 도구 설정
반복적인 확인 작업을 줄이기 위해 코드 검사를 자동 실행하는 기본 도구 설정.
- 이 코드를 동료 개발자가 쉽게 이해하고 따라갈 수 있을까?
- 업무 도메인에 대해 말하고 있는가?
- 팀에 새로 합류하는 사람도 쉽게 이해하고 효과적으로 작업할 수 있을까?    


이 모든 검사는 자동화 되어야 함.
### Mypy를 사용한 타입 힌팅
- [mypy](http://mypy-lang.org/) 는 파이썬에서 가장 일반적으로 사용하는 정적 타입 검사도구.
- 프로젝트의 모든 파일을 분석해 타입 불일치를 검사.
- `pip install mypy`
- 잘못된 탐지를 할 경우 문장 끝에 주석을 추가해 mypy 무시되도록 할 수 있음
    - `type_to_ignore = "something" # type: ignore`

### pylint를 사용한 코드 검사
- `pip install pylint`
- .pylintrc 파일을 통해 설정 값 바꿀 수 있음
     - 이 파일을 통해 규칙 활성화 또는 비활성화 가능. 한 줄의 최대 글자 수 같은 값 설정 가능
     
### 자동 검사 설정
리눅스 개발환경에서 빌드 자동화 하는 가장 일반적 방법은 **makefile** 사용하는 것
- 프로젝트 컴파일하고 실행하기 위한 설정 도와주는 파워풀한 도구.
- 포매팅 검사나 코딩 컨벤션 검사를 자동화하기 위해 사용할수도 있음.   

`make checklist`
- 코딩 가이드라인 검사(예를 들면 PEP-8)
- 올바른 타입을 사용했는지 검사
- 최종적으로 테스트 진행    
이 중 어떤 단계라도 실패 시 전체 프로세스가 실패한 것으로 간주.      

[Black](https://github.com/psf/black) : 자동으로 코드 포매팅.
- 문자열에 항상 쌍따옴표 사용, 파라미터 형태 항상 같은 구조 사용.
- 코드 차이를 최소화.
- PIP-8봏다 엄격한 기준이지만 도구 사용함으로써 포매팅 걱정 할 필요 없이 문제 핵심에만 집중할 수 있음.
